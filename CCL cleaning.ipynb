```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Customer Call List.xlsx")
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop(columns='Not_Useful_Column', inplace=True)
df
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Last_Name'])
df
```

```python
df.reset_index(drop=True)
```

```python
df['First_Name'] = df['First_Name'].str.upper()
df['Last_Name'] = df['Last_Name'].str.upper()
df
```

```python
df['Last_Name'] = df['Last_Name'].str.replace('[^A-Z]','',regex=True)
df['Last_Name'] = df['Last_Name'].str.strip()
df.reset_index(drop=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + '-' + x[3:6] +'-'+ x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','',regex=True)
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','',regex=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y',regex=True)
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N',regex=True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y',regex=True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N',regex=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.strip()
```

```python
df['Do_Not_Contact'].isna().sum()
```

```python
df['Do_Not_Contact'].fillna('N',inplace=True)
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('N/a','')
df['Address']=df['Address'].str.replace('N/a','')
df
```

```python
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
df
```

```python
df = df[df['Do_Not_Contact'] != 'Y']
```

```python
df.reset_index(drop=True)
```

```python
df.to_excel('Customer Call List _ Cleaned.xlsx')
```
